This notebook explores both raw and processed datasets and visualizes key patterns for the Moonshot assignment.

## 1. Import Libraries and Load Data

```python
from pathlib import Path
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio

pd.set_option('display.max_columns', 200)
pd.set_option('display.width', 200)

# Plotly renderer fallback for environments where nbformat is unavailable.
# This prevents fig.show() from failing in some notebook kernels.
try:
    import nbformat  # noqa: F401
    pio.renderers.default = 'notebook_connected'
except Exception:
    pio.renderers.default = 'vscode'

ROOT = Path.cwd()
RAW_PATH = ROOT / 'data' / 'raw' / 'amazon_luggage_raw.csv'
PROCESSED_PATH = ROOT / 'data' / 'processed' / 'amazon_luggage_processed.csv'

raw_df = pd.read_csv(RAW_PATH)
processed_df = pd.read_csv(PROCESSED_PATH)

print('Raw path      :', RAW_PATH)
print('Processed path:', PROCESSED_PATH)
print('Raw shape     :', raw_df.shape)
print('Processed shape:', processed_df.shape)
```

## 3. Core Counts and Basic Checks

```python
summary = {
    'raw_rows': len(raw_df),
    'processed_rows': len(processed_df),
    'raw_brands': raw_df['brand'].nunique(),
    'processed_brands': processed_df['brand'].nunique(),
    'raw_products': raw_df['product_title'].nunique(),
    'processed_products': processed_df['product_title'].nunique(),
    'raw_categories': raw_df['category'].nunique(),
    'processed_categories': processed_df['category'].nunique(),
}

pd.DataFrame([summary])
```

```python
print('Non-null counts (raw):')
display(raw_df.count().sort_values(ascending=False).to_frame('non_null_count'))

print('Non-null counts (processed):')
display(processed_df.count().sort_values(ascending=False).to_frame('non_null_count'))
```

```python
print('Missing values (raw):')
display(raw_df.isna().sum().to_frame('missing_count'))

print('Missing values (processed):')
display(processed_df.isna().sum().to_frame('missing_count'))

print('Duplicate rows in raw:', raw_df.duplicated().sum())
print('Duplicate rows in processed:', processed_df.duplicated().sum())
```

## 4. Descriptive Statistics

```python
print('Raw numeric summary:')
display(raw_df.describe(include=[np.number]).T)

print('Processed numeric summary:')
display(processed_df.describe(include=[np.number]).T)
```

## 5. Brand and Category Distribution

```python
brand_counts = raw_df['brand'].value_counts().reset_index()
brand_counts.columns = ['brand', 'review_rows']

fig = px.bar(brand_counts, x='brand', y='review_rows', text='review_rows',
             title='Raw Dataset Review Rows by Brand', color='brand')
fig.update_layout(showlegend=False)
fig.show()
```

```python
cat_counts = processed_df['category'].value_counts().reset_index()
cat_counts.columns = ['category', 'product_count']

fig = px.pie(cat_counts, names='category', values='product_count',
             title='Processed Dataset Category Share')
fig.show()
```

## 6. Price, Rating, Discount Relationship

```python
products_raw = raw_df.drop_duplicates(subset=['brand', 'product_title']).copy()

fig = px.scatter(
    products_raw,
    x='price',
    y='rating',
    color='brand',
    size='review_count',
    hover_data=['product_title', 'discount', 'category'],
    title='Price vs Rating (Bubble Size = Review Count)'
)
fig.show()
```

```python
fig = px.scatter(
    processed_df,
    x='discount',
    y='avg_sentiment',
    color='brand',
    size='review_count',
    hover_data=['product_title', 'price', 'rating'],
    trendline='ols',
    title='Discount vs Average Sentiment (Processed Data)'
)
fig.show()
```

## 7. Distribution Plots

```python
fig = px.histogram(
    products_raw,
    x='price',
    nbins=25,
    color='brand',
    barmode='overlay',
    opacity=0.6,
    title='Price Distribution by Brand'
)
fig.show()
```

```python
fig = px.box(
    products_raw,
    x='brand',
    y='rating',
    color='brand',
    title='Rating Spread by Brand'
)
fig.update_layout(showlegend=False)
fig.show()
```

## 8. Correlation Heatmap (Processed Numeric Features)

```python
numeric_cols = processed_df.select_dtypes(include=[np.number]).columns.tolist()
corr = processed_df[numeric_cols].corr(numeric_only=True)

fig = px.imshow(
    corr,
    text_auto='.2f',
    aspect='auto',
    color_continuous_scale='RdBu_r',
    title='Correlation Matrix - Processed Dataset'
)
fig.show()
```

## 9. Top and Bottom Products by Sentiment

```python
top_sent = processed_df.sort_values('avg_sentiment', ascending=False).head(10)
bottom_sent = processed_df.sort_values('avg_sentiment', ascending=True).head(10)

display(top_sent[['brand', 'product_title', 'avg_sentiment', 'price', 'rating', 'discount']])
display(bottom_sent[['brand', 'product_title', 'avg_sentiment', 'price', 'rating', 'discount']])
```

## 10. Key Takeaways Helper Cell
Use this cell to print a few quick summary statements for presentation/demo.

```python
best_brand_sent = processed_df.groupby('brand')['avg_sentiment'].mean().sort_values(ascending=False).index[0]
cheapest_brand = processed_df.groupby('brand')['price'].mean().sort_values().index[0]
highest_discount_brand = processed_df.groupby('brand')['discount'].mean().sort_values(ascending=False).index[0]

print(f'Best average sentiment brand: {best_brand_sent}')
print(f'Most affordable average-price brand: {cheapest_brand}')
print(f'Highest average discount brand: {highest_discount_brand}')
```
